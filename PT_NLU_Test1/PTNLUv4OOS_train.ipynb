```python
class CBv2Dataset(Dataset):
    """Dataset class for RoBERTa intent classification with optimized encoding"""
    
    def __init__(self, texts: List[str], main_intents: List[str], 
                 sub_intents: List[str], tokenizer: Any, max_len: int = 128,
                 precompute_encodings: bool = True):
        self.texts = texts
        self.main_intents = main_intents
        self.sub_intents = sub_intents
        self.tokenizer = tokenizer
        self.max_len = max_len
        self.encodings = None
        
        # Pre-compute encodings for all texts in batches (much faster)
        if precompute_encodings:
            logger.info("Pre-computing encodings for all texts (this will speed up training)")
            self._precompute_encodings()
    
    def _precompute_encodings(self):
        """Pre-compute encodings in batches to save time during training"""
        batch_size = 256  # Process this many texts at once
        batches = [self.texts[i:i + batch_size] for i in range(0, len(self.texts), batch_size)]
        
        # Prepare to store the encodings
        self.encodings = {
            'input_ids': [],
            'attention_mask': []
        }
        
        # Process each batch with tqdm for progress
        logger.info(f"Processing {len(batches)} batches of encodings")
        for i, batch_texts in enumerate(tqdm(batches, desc="Encoding batches")):
            # Encode the batch
            batch_encodings = self.tokenizer.batch_encode_plus(
                batch_texts,
                add_special_tokens=True,
                max_length=self.max_len,
                padding='max_length',
                truncation=True,
                return_attention_mask=True,
                return_tensors='pt'
            )
            
            # Store the encodings
            self.encodings['input_ids'].append(batch_encodings['input_ids'])
            self.encodings['attention_mask'].append(batch_encodings['attention_mask'])
            
            # Clear GPU cache if using CUDA
            if torch.cuda.is_available():
                torch.cuda.empty_cache()
        
        # Concatenate all batches
        self.encodings['input_ids'] = torch.cat(self.encodings['input_ids'], dim=0)
        self.encodings['attention_mask'] = torch.cat(self.encodings['attention_mask'], dim=0)
        
        logger.info(f"Completed encoding {len(self.texts)} texts")

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        try:
            text = str(self.texts[idx])
            main_intent = self.main_intents[idx]
            sub_intent = self.sub_intents[idx]
            
            # Use pre-computed encodings if available
            if self.encodings is not None:
                return {
                    'text': text,
                    'input_ids': self.encodings['input_ids'][idx],
                    'attention_mask': self.encodings['attention_mask'][idx],
                    'main_intent': torch.tensor(main_intent, dtype=torch.long),
                    'sub_intent': torch.tensor(sub_intent, dtype=torch.long)
                }
            
            # Fallback to on-the-fly encoding (slower)
            encoding = self.tokenizer.encode_plus(
                text,
                add_special_tokens=True,
                max_length=self.max_len,
                padding='max_length',
                truncation=True,
                return_attention_mask=True,
                return_tensors='pt'
            )
            
            return {
                'text': text,
                'input_ids': encoding['input_ids'].flatten(),
                'attention_mask': encoding['attention_mask'].flatten(),
                'main_intent': torch.tensor(main_intent, dtype=torch.long),
                'sub_intent': torch.tensor(sub_intent, dtype=torch.long)
            }
        except Exception as e:
            logger.error(f"Error processing dataset item at index {idx}: {str(e)}")
            raise
```

```python
class HierarchicalRobertaClassifier(nn.Module):
    """Hierarchical RoBERTa classifier with OOS detection"""
    
    def __init__(self, intent_structure: Dict[str, Dict[str, List[str]]], dropout_rate: float = 0.3):
        super().__init__()
        try:
            # Use RoBERTa instead of BERT
            self.roberta = RobertaModel.from_pretrained('roberta-base')
            self.dropout = nn.Dropout(dropout_rate)
            
            # Main intent classifier (including OOS)
            self.n_main_intents = len(intent_structure)
            self.main_classifier = nn.Linear(768, self.n_main_intents)  # RoBERTa-base has 768 dim
            
            # Create sub-classifiers for each main intent (excluding OOS)
            self.sub_classifiers = nn.ModuleDict()
            for main_intent, sub_intents in intent_structure.items():
                if main_intent != 'oos':  # OOS doesn't have sub-intents
                    self.sub_classifiers[main_intent] = nn.Linear(768, len(sub_intents))
            
            # Store the structure for reference
            self.intent_structure = intent_structure
            
        except Exception as e:
            logger.error(f"Error initializing hierarchical RoBERTa classifier: {str(e)}")
            raise

    def forward(self, input_ids, attention_mask, main_intent=None):
        try:
            # RoBERTa forward pass
            outputs = self.roberta(input_ids=input_ids, attention_mask=attention_mask)
            # Use pooler output (CLS token representation)
            pooled_output = outputs.pooler_output
            dropped = self.dropout(pooled_output)
            
            # Get main intent prediction
            main_output = self.main_classifier(dropped)
            
            if main_intent is None:
                main_intent = torch.argmax(main_output, dim=1)
            
            # Get sub-intent prediction for the specific main intent
            batch_size = input_ids.size(0)
            max_sub_intents = max(len(subs) for main, subs in self.intent_structure.items() if main != 'oos')
            sub_outputs = torch.zeros(batch_size, max_sub_intents).to(input_ids.device)
            
            for i in range(batch_size):
                main_intent_names = list(self.intent_structure.keys())
                intent_name = main_intent_names[main_intent[i]]
                
                # Skip sub-classification for OOS
                if intent_name == 'oos':
                    continue
                    
                if intent_name in self.sub_classifiers:
                    sub_classifier = self.sub_classifiers[intent_name]
                    sub_output = sub_classifier(dropped[i].unsqueeze(0))
                    sub_outputs[i, :sub_output.size(1)] = sub_output
                
            return main_output, sub_outputs
            
        except Exception as e:
            logger.error(f"Error in forward pass: {str(e)}")
            raise
```

```python
def create_sample_training_data():
    """Create sample training data with business intents + OOS data"""
    
    # # Sample business intent structure
    # business_intent_structure = {
    #     "order_management": {
    #         "create_order": ["I want to place an order", "create new order", "place order"],
    #         "cancel_order": ["cancel my order", "remove order", "delete order"],
    #         "track_order": ["where is my order", "track order status", "order tracking"]
    #     },
    #     "product_inquiry": {
    #         "product_info": ["tell me about product", "product details", "what is this product"],
    #         "pricing": ["what's the price", "how much does it cost", "pricing information"],
    #         "availability": ["is this available", "product in stock", "availability check"]
    #     },
    #     "customer_support": {
    #         "technical_issue": ["not working", "system error", "technical problem"],
    #         "billing_issue": ["billing problem", "invoice error", "payment issue"],
    #         "general_help": ["need help", "customer support", "assistance required"]
    #     },
    #     "oos": {
    #         "out_of_scope": []  # Will be populated with OOS data
    #     }
    # }
    
    # Prepare training data
    texts = []
    main_intents = []
    sub_intents = []
    
    # Add business intents
    main_intent_to_id = {}
    sub_intent_to_id = {}
    
    for main_idx, (main_intent, sub_intent_dict) in enumerate(intent_samples.items()):
        main_intent_to_id[main_intent] = main_idx
        sub_intent_to_id[main_intent] = {}
        
        for sub_idx, (sub_intent, examples) in enumerate(sub_intent_dict.items()):
            sub_intent_to_id[main_intent][sub_intent] = sub_idx
            
            # Add examples for this sub-intent
            for example in examples:
                texts.append(example)
                main_intents.append(main_idx)
                sub_intents.append(sub_idx)
    
    # Add OOS data
    oos_main_id = main_intent_to_id['oos']
    oos_sub_id = sub_intent_to_id['oos']['out_of_scope']
    
    for category, examples in OOS_TRAINING_DATA.items():
        for example in examples:
            texts.append(example)
            main_intents.append(oos_main_id)
            sub_intents.append(oos_sub_id)
    
    logger.info(f"Created sample dataset with {len(texts)} examples")
    logger.info(f"Business intents: {len([t for t in texts if main_intents[texts.index(t)] != oos_main_id])}")
    logger.info(f"OOS examples: {len([t for t in texts if main_intents[texts.index(t)] == oos_main_id])}")
    
    return texts, main_intents, sub_intents, intent_samples, main_intent_to_id, sub_intent_to_id
```

```python
def train_model(model, train_loader, val_loader, device, epochs=5, learning_rate=2e-5):
    """Train the hierarchical RoBERTa model"""
    
    optimizer = AdamW(model.parameters(), lr=learning_rate, weight_decay=0.01)
    
    # Calculate total steps for scheduler
    total_steps = len(train_loader) * epochs
    scheduler = get_linear_schedule_with_warmup(
        optimizer,
        num_warmup_steps=int(0.1 * total_steps),
        num_training_steps=total_steps
    )
    
    # Loss functions
    main_criterion = nn.CrossEntropyLoss()
    sub_criterion = nn.CrossEntropyLoss()
    
    model.train()
    
    training_stats = []
    
    for epoch in range(epochs):
        logger.info(f"Starting Epoch {epoch + 1}/{epochs}")
        
        total_train_loss = 0
        main_correct = 0
        sub_correct = 0
        total_samples = 0
        
        # Training loop
        progress_bar = tqdm(train_loader, desc=f"Epoch {epoch + 1}")
        
        for batch in progress_bar:
            # Move to device
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            main_labels = batch['main_intent'].to(device)
            sub_labels = batch['sub_intent'].to(device)
            
            optimizer.zero_grad()
            
            # Forward pass
            main_outputs, sub_outputs = model(input_ids, attention_mask)
            
            # Calculate losses
            main_loss = main_criterion(main_outputs, main_labels)
            
            # Only calculate sub-loss for non-OOS samples
            non_oos_mask = main_labels != 3  # Assuming OOS is index 3
            if non_oos_mask.sum() > 0:
                sub_loss = sub_criterion(sub_outputs[non_oos_mask], sub_labels[non_oos_mask])
                total_loss = main_loss + 0.5 * sub_loss  # Weight sub-loss lower
            else:
                total_loss = main_loss
            
            # Backward pass
            total_loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            scheduler.step()
            
            # Statistics
            total_train_loss += total_loss.item()
            main_correct += (torch.argmax(main_outputs, dim=1) == main_labels).sum().item()
            if non_oos_mask.sum() > 0:
                sub_correct += (torch.argmax(sub_outputs[non_oos_mask], dim=1) == sub_labels[non_oos_mask]).sum().item()
            total_samples += input_ids.size(0)
            
            # Update progress bar
            progress_bar.set_postfix({
                'Loss': f'{total_loss.item():.4f}',
                'Main Acc': f'{main_correct/total_samples:.4f}',
                'LR': f'{scheduler.get_last_lr()[0]:.2e}'
            })
        
        # Validation
        val_main_acc, val_sub_acc, val_loss = evaluate_model(model, val_loader, device)
        
        # Log epoch results
        epoch_stats = {
            'epoch': epoch + 1,
            'train_loss': total_train_loss / len(train_loader),
            'train_main_acc': main_correct / total_samples,
            'val_main_acc': val_main_acc,
            'val_sub_acc': val_sub_acc,
            'val_loss': val_loss
        }
        training_stats.append(epoch_stats)
        
        logger.info(f"Epoch {epoch + 1} - Train Loss: {epoch_stats['train_loss']:.4f}, "
                   f"Val Main Acc: {val_main_acc:.4f}, Val Sub Acc: {val_sub_acc:.4f}")
    
    return training_stats
```

```python
def evaluate_model(model, val_loader, device):
    """Evaluate the model on validation data"""
    model.eval()
    
    main_criterion = nn.CrossEntropyLoss()
    sub_criterion = nn.CrossEntropyLoss()
    
    total_val_loss = 0
    main_correct = 0
    sub_correct = 0
    non_oos_samples = 0
    total_samples = 0
    
    with torch.no_grad():
        for batch in val_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            main_labels = batch['main_intent'].to(device)
            sub_labels = batch['sub_intent'].to(device)
            
            main_outputs, sub_outputs = model(input_ids, attention_mask)
            
            # Calculate losses
            main_loss = main_criterion(main_outputs, main_labels)
            non_oos_mask = main_labels != 3  # Assuming OOS is index 3
            
            if non_oos_mask.sum() > 0:
                sub_loss = sub_criterion(sub_outputs[non_oos_mask], sub_labels[non_oos_mask])
                total_loss = main_loss + 0.5 * sub_loss
                sub_correct += (torch.argmax(sub_outputs[non_oos_mask], dim=1) == sub_labels[non_oos_mask]).sum().item()
                non_oos_samples += non_oos_mask.sum().item()
            else:
                total_loss = main_loss
            
            total_val_loss += total_loss.item()
            main_correct += (torch.argmax(main_outputs, dim=1) == main_labels).sum().item()
            total_samples += input_ids.size(0)
    
    model.train()
    
    main_acc = main_correct / total_samples
    sub_acc = sub_correct / non_oos_samples if non_oos_samples > 0 else 0
    avg_loss = total_val_loss / len(val_loader)
    
    return main_acc, sub_acc, avg_loss
```

```python
def save_model(model, tokenizer, intent_structure, main_intent_to_id, sub_intent_to_id, save_path):
    """Save the trained model and associated data"""
    
    # Create save directory if it doesn't exist
    os.makedirs(os.path.dirname(save_path), exist_ok=True)
    
    # Prepare checkpoint
    checkpoint = {
        'model_state_dict': model.state_dict(),
        'intent_structure': intent_structure,
        'main_intent_to_id': main_intent_to_id,
        'sub_intent_to_id': sub_intent_to_id,
        'model_config': {
            'model_type': 'roberta-base',
            'n_main_intents': len(intent_structure),
            'dropout_rate': 0.3
        },
        'timestamp': datetime.now().isoformat()
    }
    
    # Save checkpoint
    torch.save(checkpoint, save_path)
    logger.info(f"Model saved to {save_path}")
```

```python
class RobertaIntentClassifier:
    """Inference class for the trained RoBERTa model"""
    
    def __init__(self, model_path: str, confidence_threshold: float = 0.7):
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.tokenizer = RobertaTokenizer.from_pretrained('roberta-base')
        self.confidence_threshold = confidence_threshold
        
        # Load model
        self.load_model(model_path)
    
    def load_model(self, model_path: str):
        """Load the trained model"""
        checkpoint = torch.load(model_path, map_location=self.device)
        
        self.intent_structure = checkpoint['intent_structure']
        self.main_intent_to_id = checkpoint['main_intent_to_id']
        self.sub_intent_to_id = checkpoint['sub_intent_to_id']
        
        # Create model
        self.model = HierarchicalRobertaClassifier(self.intent_structure)
        self.model.load_state_dict(checkpoint['model_state_dict'])
        self.model.to(self.device)
        self.model.eval()
        
        logger.info(f"Model loaded successfully from {model_path}")
    
    def predict(self, text: str) -> Dict[str, Any]:
        """Predict intent for given text"""
        if not text or not text.strip():
            return {
                'status': 'error',
                'error': 'Empty input text',
                'confidence': {'main': 0.0, 'sub': 0.0}
            }
        
        try:
            # Tokenize
            encoding = self.tokenizer.encode_plus(
                text,
                add_special_tokens=True,
                max_length=128,
                padding='max_length',
                truncation=True,
                return_attention_mask=True,
                return_tensors='pt'
            )
            
            input_ids = encoding['input_ids'].to(self.device)
            attention_mask = encoding['attention_mask'].to(self.device)
            
            with torch.no_grad():
                main_outputs, sub_outputs = self.model(input_ids, attention_mask)
                
                # Get main intent prediction
                main_probs = torch.softmax(main_outputs, dim=1)
                main_confidence, main_predicted = torch.max(main_probs, 1)
                
                # Convert IDs back to intent names
                id_to_main_intent = {v: k for k, v in self.main_intent_to_id.items()}
                predicted_main_intent = id_to_main_intent[main_predicted.item()]
                
                # Check if it's OOS
                if predicted_main_intent == 'oos':
                    return {
                        'status': 'oos_detected',
                        'main_intent': 'oos',
                        'sub_intent': 'out_of_scope',
                        'confidence': {
                            'main': main_confidence.item(),
                            'sub': 0.0
                        },
                        'should_fallback': True
                    }
                
                # Check confidence threshold for business intents
                if main_confidence.item() < self.confidence_threshold:
                    return {
                        'status': 'low_confidence',
                        'confidence': {
                            'main': main_confidence.item(),
                            'sub': 0.0
                        },
                        'should_fallback': True,
                        'predicted_intent': predicted_main_intent
                    }
                
                # Get sub-intent prediction
                sub_probs = torch.softmax(sub_outputs, dim=1)
                sub_confidence, sub_predicted = torch.max(sub_probs, 1)
                
                id_to_sub_intent = {v: k for k, v in self.sub_intent_to_id[predicted_main_intent].items()}
                predicted_sub_intent = id_to_sub_intent[sub_predicted.item()]
                
                return {
                    'status': 'success',
                    'main_intent': predicted_main_intent,
                    'sub_intent': predicted_sub_intent,
                    'confidence': {
                        'main': main_confidence.item(),
                        'sub': sub_confidence.item()
                    },
                    'should_fallback': False
                }
                
        except Exception as e:
            logger.error(f"Error during prediction: {str(e)}")
            return {
                'status': 'error',
                'error': str(e),
                'confidence': {'main': 0.0, 'sub': 0.0}
            }
```

```python
class BinaryOOSDetector(nn.Module):
    """Binary classifier to detect OOS vs Business intents"""
    
    def __init__(self, dropout_rate: float = 0.3):
        super().__init__()
        self.roberta = RobertaModel.from_pretrained('roberta-base')
        self.dropout = nn.Dropout(dropout_rate)
        self.classifier = nn.Linear(768, 2)  # Binary: OOS vs Business
        
    def forward(self, input_ids, attention_mask):
        outputs = self.roberta(input_ids=input_ids, attention_mask=attention_mask)
        pooled_output = outputs.pooler_output
        dropped = self.dropout(pooled_output)
        return self.classifier(dropped)

def train_binary_oos_detector(texts, labels, tokenizer, device, epochs=3):
    """Train binary OOS detector first"""
    
    # Create binary labels (0=Business, 1=OOS)
    binary_labels = [1 if label == 3 else 0 for label in labels]  # Assuming OOS is class 3
    
    # Create dataset
    dataset = CBv2Dataset(texts, binary_labels, [0]*len(texts), tokenizer)
    dataloader = DataLoader(dataset, batch_size=16, shuffle=True)
    
    # Initialize binary model
    binary_model = BinaryOOSDetector()
    binary_model.to(device)
    
    optimizer = AdamW(binary_model.parameters(), lr=2e-5)
    criterion = nn.CrossEntropyLoss()
    
    binary_model.train()
    
    for epoch in range(epochs):
        total_loss = 0
        correct = 0
        total = 0
        
        for batch in tqdm(dataloader, desc=f"Binary Training Epoch {epoch+1}"):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['main_intent'].to(device)
            
            optimizer.zero_grad()
            outputs = binary_model(input_ids, attention_mask)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
            predicted = torch.argmax(outputs, dim=1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)
        
        accuracy = correct / total
        logger.info(f"Binary Epoch {epoch+1}: Loss={total_loss/len(dataloader):.4f}, Acc={accuracy:.4f}")
    
    return binary_model
    """Train the hierarchical RoBERTa model"""
    
    optimizer = AdamW(model.parameters(), lr=learning_rate, weight_decay=0.01)
    
    # Calculate total steps for scheduler
    total_steps = len(train_loader) * epochs
    scheduler = get_linear_schedule_with_warmup(
        optimizer,
        num_warmup_steps=int(0.1 * total_steps),
        num_training_steps=total_steps
    )
    
    # Calculate class weights for main intents (heavily weight OOS)
    oos_id = main_intent_to_id.get('oos', 3)  # Get OOS class ID
    n_classes = len(intent_structure)
    
    # Create class weights - give OOS 5x more weight
    class_weights = torch.ones(n_classes)
    class_weights[oos_id] = 5.0  # Heavy weight for OOS class
    class_weights = class_weights.to(device)
    
    # Loss functions with class weights
    main_criterion = nn.CrossEntropyLoss(weight=class_weights)
    sub_criterion = nn.CrossEntropyLoss()
    
    model.train()
    
    training_stats = []
    
    for epoch in range(epochs):
        logger.info(f"Starting Epoch {epoch + 1}/{epochs}")
        
        total_train_loss = 0
        main_correct = 0
        sub_correct = 0
        total_samples = 0
        
        # Training loop
        progress_bar = tqdm(train_loader, desc=f"Epoch {epoch + 1}")
        
        for batch in progress_bar:
            # Move to device
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            main_labels = batch['main_intent'].to(device)
            sub_labels = batch['sub_intent'].to(device)
            
            optimizer.zero_grad()
            
            # Forward pass
            main_outputs, sub_outputs = model(input_ids, attention_mask)
            
            # Calculate losses
            main_loss = main_criterion(main_outputs, main_labels)
            
            # Only calculate sub-loss for non-OOS samples
            non_oos_mask = main_labels != 3  # Assuming OOS is index 3
            if non_oos_mask.sum() > 0:
                sub_loss = sub_criterion(sub_outputs[non_oos_mask], sub_labels[non_oos_mask])
                total_loss = main_loss + 0.5 * sub_loss  # Weight sub-loss lower
            else:
                total_loss = main_loss
            
            # Backward pass
            total_loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            scheduler.step()
            
            # Statistics
            total_train_loss += total_loss.item()
            main_correct += (torch.argmax(main_outputs, dim=1) == main_labels).sum().item()
            if non_oos_mask.sum() > 0:
                sub_correct += (torch.argmax(sub_outputs[non_oos_mask], dim=1) == sub_labels[non_oos_mask]).sum().item()
            total_samples += input_ids.size(0)
            
            # Update progress bar
            progress_bar.set_postfix({
                'Loss': f'{total_loss.item():.4f}',
                'Main Acc': f'{main_correct/total_samples:.4f}',
                'LR': f'{scheduler.get_last_lr()[0]:.2e}'
            })
        
        # Validation
        val_main_acc, val_sub_acc, val_loss = evaluate_model(model, val_loader, device)
        
        # Log epoch results
        epoch_stats = {
            'epoch': epoch + 1,
            'train_loss': total_train_loss / len(train_loader),
            'train_main_acc': main_correct / total_samples,
            'val_main_acc': val_main_acc,
            'val_sub_acc': val_sub_acc,
            'val_loss': val_loss
        }
        training_stats.append(epoch_stats)
        
        logger.info(f"Epoch {epoch + 1} - Train Loss: {epoch_stats['train_loss']:.4f}, "
                   f"Val Main Acc: {val_main_acc:.4f}, Val Sub Acc: {val_sub_acc:.4f}")
    
    return training_stats
```
